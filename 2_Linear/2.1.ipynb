```julia
using FundamentalsNumericalComputation
```

```julia
#given polynomial, plot the graph
function plot_interpolation(p, a, b, n = 1000)
    tt = range(a,b, length = n)
    yy = p.(tt)
    plot!(tt, yy, label = "Interpolant")
end
```

```julia
#1 
points = [(-1, 0), (0,1), (2,0), (3, 1), (4, 2)]
x = [point[1] for point in points]
y = [point[2] for point in points]
V = [x[i]^(j) for i in 1:5, j in 0:4]
c = V\y 
p = Polynomial(c)
scatter(x, y, label="data points")
plot_interpolation(p, -1, 4)
```
